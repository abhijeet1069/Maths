# Vector Equations

A vector means an ordered list of numbers.

Sometimes for convinience 

$
\begin{bmatrix}
 3 \\
-1
\end{bmatrix}
$
can be written in the form (3,-1)

Also,
$
\begin{bmatrix}
 3 \\
-1
\end{bmatrix}
\neq
\begin{bmatrix}
 3 & -1
\end{bmatrix}
$


```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
vectors = np.array([[2, 2], [-6, 1], [-4, 3]])  # Each row is a vector

# Origin (0,0) for all vectors
origin = np.zeros((vectors.shape[0], 2))

# Plot vectors
plt.quiver(*origin.T, vectors[:, 0], vectors[:, 1], angles='xy', scale_units='xy', scale=1, color=['r', 'g', 'b'])

# Set axis limits
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Labels and title
plt.grid()
plt.title("2D Vectors")
plt.show()
```

## Linear Combinations

Given vectors $ v_1, v_2, ... ,v_p $ in $R^n$ and given scalars(weights) $c_1, c_2,...,c_p$ the vector y is defined by

$ y = c_1v_1 + ... c_pv_p$ is a linear combination.

## Span

$ Span \{v_1, v_2, ... ,v_p\} $ is the collection of all vectors that can be written in the form 
$ c_1v_1 + c_2v_2 + ...+ c_pv_p$

Asking whether a vector b is in the $ Span \{v_1, v_2, ... ,v_p\} $ amounts to asking whether the vector equation 
$ x_1v_1 + x_2v_2 + ...+ x_pv_p = b $
