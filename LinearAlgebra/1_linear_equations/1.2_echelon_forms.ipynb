# Row reduction and Echelon forms

## Elementary row operations

- Replace one row by the sum of itself and multiple of another row.
- Interchange rows.
- Multiply all entries in a row by a non-zero constant

## Echelon Form

A rectangular matrix is in echelon form (or row echelon form) if it has below 3 properties:

1. All zero rows (if any) are at the bottom.
2. The first nonzero element (called a pivot) in each row is to the right of the pivot in the row above.
3. The pivot (leading coefficient) in a nonzero row is **nonzero**.

If a matrix satisfies below additional conditions then it is in reduced echelon form(or reduced row echelon form):

4. Every pivot is **1**.
5. Each pivot is the **only nonzero entry** in its column.

Row echelon form : 

$
\begin{bmatrix}
1 & 2 & 3 & 4 \\
0 & 5 & 6 & 7 \\
0 & 0 & 8 & 9
\end{bmatrix}
$

Reduced Echelon Form : 

$
\begin{bmatrix}
1 & 0 & 2 & 3 \\
0 & 1 & 4 & 5 \\
0 & 0 & 0 & 1
\end{bmatrix}
$

## Uniqueness of the Reduced Echelon form

Each matrix is row equivalent to one and only one reduced echelon matrix.

## Pivot Positions

A pivot position in a matrix A is a locaion in A that corresponds to a leading 1 in the reduced echelon form of A. A 
pivot column is a column of A that contains a pivot position.

## Solutions of Linear systems

In reduced echelon form pivot positions are basic variables, while other variables are called free variable.
