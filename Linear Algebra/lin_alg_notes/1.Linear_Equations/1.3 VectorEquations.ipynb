# Vector Equations

A vector can be imagined as a movement.
A matrix with only one column is called is called a column vector, or simply a vector.

$ 
u = \begin{bmatrix} 
 1 \\ 
 0 \\
 5  
 \end{bmatrix}
$

We use this notation (1,0,5) also for column vector, to save space. Note, this is not equal to [1,0,5]
as both are completely different shape.

![Vector Algebra](img/vector-algebra.png)

## Linear Combinations

Given vectors $ v_1, v_2,...,v_p $ in $\Reals^n$ and given scalars $c_1, c_2,..,c_p$ the vector y is defined as
$y = c_1v_1 +...+ c_pv_p $
is called a linear combination of $ v_1,..,v_p $ with weights $ c_1,..,c_p $
 

## Span

Span $(v_1,..,v_p)$ is a collection of all vectors that can be written in the form,
$c_1.v_1 + c_2.v_2 + ... +c_p.v_p $ with $ c_1,..,c_p $ scalars

Or, simply span of a vector is a set of all linear combinations of a vector.


```python
# Span of 1 vector
import numpy as np
import matplotlib.pyplot as plt

# Define the vector
v = np.array([3, 9])

# Generate points along the span of the vector
t = np.linspace(-10, 10, 100)  # Range of scalar multiples
span_points = t[:, np.newaxis] * v  # Linear combinations

# Plot the vector and its span
plt.figure(figsize=(8, 8))

# Plot the span line
plt.plot(span_points[:, 0], span_points[:, 1], 'b-', linewidth=1, label='Span of vector')

# Plot the vector itself
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector v')

# Setting up the plot
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.xlim(-40, 40)
plt.ylim(-40, 40)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')

# Labels and legend
plt.title("Span of a Vector in 2D")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()

plt.show()
```

```python
# Span of 2d vectors
import numpy as np
import matplotlib.pyplot as plt

# Define the two vectors
v1 = np.array([3, 2])
v2 = np.array([4, -4])

# Create a grid of coefficients for linear combinations
coeff = np.linspace(-10, 10, 21)
x, y = np.meshgrid(coeff, coeff)

# Compute all linear combinations
span_points = x[..., np.newaxis] * v1 + y[..., np.newaxis] * v2

# Plot the span
plt.figure(figsize=(8, 8))

# Draw the span points
for i in range(span_points.shape[0]):
    for j in range(span_points.shape[1]):
        plt.plot(span_points[i, j, 0], span_points[i, j, 1], 'k.', markersize=2)

# Plot the vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='v2')

# Customize the plot
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.xlim(-30, 30)
plt.ylim(-30, 30)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Span of Two Vectors in 2D")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource

# Define the two vectors
v1 = np.array([1, 0, 0])
v2 = np.array([0, 1, 0])

# Create a grid of coefficients for linear combinations
coeff = np.linspace(-2, 2, 50)  # Increased grid resolution
u, v = np.meshgrid(coeff, coeff)

# Compute all points in the span of v1 and v2
plane = u[..., np.newaxis] * v1 + v[..., np.newaxis] * v2

# Extract x, y, z coordinates
X, Y, Z = plane[..., 0], plane[..., 1], plane[..., 2]

# Plotting
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Add lighting for better visualization
ls = LightSource(azdeg=225, altdeg=45)
plane_color = ls.shade(Z, cmap=plt.cm.coolwarm, blend_mode='soft')

# Plot the plane with smooth shading
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=plane_color, alpha=0.7, edgecolor='gray')

# Plot the vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='red', linewidth=2, label='v1', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='blue', linewidth=2, label='v2', arrow_length_ratio=0.1)

# Mark the origin for better visibility
ax.scatter(0, 0, 0, color='black', s=100, label='Origin')

# Plot the 3D axes
axis_limit = 3
ax.quiver(0, 0, 0, axis_limit, 0, 0, color='black', linewidth=1, arrow_length_ratio=0.05, label='X-axis')
ax.quiver(0, 0, 0, 0, axis_limit, 0, color='green', linewidth=1, arrow_length_ratio=0.05, label='Y-axis')
ax.quiver(0, 0, 0, 0, 0, axis_limit, color='purple', linewidth=1, arrow_length_ratio=0.05, label='Z-axis')

# Set axis limits and make the origin visible
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

# Add labels and title
ax.set_xlabel('X-axis', labelpad=15)
ax.set_ylabel('Y-axis', labelpad=15)
ax.set_zlabel('Z-axis', labelpad=15)
ax.set_title('Span of Two Vectors in 3D with Centered Origin and Axes', pad=20)
ax.legend()

# Improve the aspect ratio and grid
ax.set_box_aspect([1, 1, 1])  # Equal aspect ratio
ax.grid(True)

plt.show()
```
